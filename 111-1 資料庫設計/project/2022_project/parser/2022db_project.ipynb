```python
import requests
from bs4 import BeautifulSoup
import math
headers={"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36"}
link = requests.get('https://www.ss.net.tw/')
soup = BeautifulSoup(link.text, 'html.parser')

category=soup.select("h5.mb-1 a.flex-grow-1")
tag_link=[]
category_link=[]
category_title=[]
category_page=[]
for i in category:
  #print(i)#觀察出所有的tag
  category_link.append(i['href']) #儲存所有的tag_link
  category_title.append(i['title'])
  #print(math.ceil(int(i.small.text.strip('()'))/24))
  category_page.append(math.ceil(int(i.small.text.strip('()'))/24))#計算頁數
#print(len(category))
```

```python
for i in range(0,len(category)):
  print(category_title[i])
  print(category_link[i])
  print(str(category_page[i])+"頁")
```

```python
s_link=category_link[0]#選取人物、風景
```

```python
link = requests.get(s_link)
soup = BeautifulSoup(link.text, 'html.parser')
s_category=soup.select('a.painter_img')
for i in s_category:
  print(i)
#print(ss_category)
```

```python
s_category_link=[]
s_category_title=[]
s_category_image=[]
for i in s_category:
  s_category_link.append(i['href'])
  s_category_title.append(i['title'])
for i in range(0,len(s_category)):
  print(s_category_link[i])
  print(s_category_title[i])
print(s_category_title)
```

```python
s_category_page=[]
for i in range(0,len(s_category)):
  link = requests.get(s_category_link[i])
  soup = BeautifulSoup(link.text, 'html.parser')
  page=soup.select('div.NavBar small')[0].text.strip('(')
  print(page)
  s_category_page.append(math.ceil(int(page)/24))
print(s_category_page)
```

```python
templist=[]
linklist=[]
for i in range(0,len(s_category)):
  for j in range(0,s_category_page[i]):
    #print(s_category_link[i]+"~"+str(j+1))
    templist.append(s_category_link[i]+"~"+str(j+1))
  linklist.append(templist)
  templist=[]
for i in linklist:
  print(i)
#print(len(linklist))
```

```python
painting_link=[]
painting_link_list=[]
for i in linklist:
  for j in i:
    link = requests.get(j)
    soup = BeautifulSoup(link.text, 'html.parser')
    paint=soup.select('div.card a')
    #print(paint)
    for k in paint:
      painting_link.append(k['href'])
  painting_link_list.append(painting_link)
  painting_link=[]
```

```python
sum=0
for i in range(0,len(painting_link_list)):
  for j in range(0,len(painting_link_list[i])):
    print(painting_link_list[i][j])
  sum+=len(painting_link_list[i])
#print(sum)
```

```python
painting_info=[]
painting_info_list=[]
time_set=set()
painter_set=set()
tag_set=set()
pigment_set=set()
for i in range(0,len(painting_link_list)):
  for j in range(0,len(painting_link_list[i])):
    link = requests.get(painting_link_list[i][j])
    soup = BeautifulSoup(link.text, 'html.parser')
    info=soup.find('div',class_='col-md-7')
    list_unstyled=info.find('ul',class_='list-unstyled')
    li=list_unstyled.find_all('li')
    #有的作品沒有年份，所以只留有年份的當作資料
    if(len(li)==6):
      #en=info.find_all('small')
      #name_en=en[0].text
      #name_cn=info.select('h1')[0].text.strip(name_en)
      name=info.select('h1')[0].text
      label=list_unstyled.find('li').text.strip("編　　號")
      #painter_en=en[1].text
      #painter_cn=list_unstyled.find('li',class_='d-flex').find('span').text.strip(painter_en)
      painter=list_unstyled.find('li',class_='d-flex').find('span').text
      painter_set.add(painter)
      #pigment_en=en[2].text
      #pigment_cn=list_unstyled.find_all('li')[4].text.strip('原作材質').strip(pigment_en)
      pigment=list_unstyled.find_all('li')[4].text.strip('原作材質')
      pigment_set.add(pigment)
      list_inline=soup.find_all(class_='list-inline-item')
      tag=''
      for k in range(1,len(list_inline)):
        #tag_list.append(list_inline[k].text)
        tag=tag+list_inline[k].text
        tag_set.add(list_inline[k].text)
      time=soup.find_all(class_='breadcrumb-item')[1].text
      time_set.add(time)
      attr=s_category_title[i]
      img=soup.find(class_="col-md-5 col-lg-6").find('img')['data-src']


      painting_info.append(time)
      #painting_info.append(name_cn)
      #painting_info.append(name_en)
      painting_info.append(label)
      painting_info.append(name)
      #painting_info.append(painter_cn)
      #painting_info.append(painter_en)
      painting_info.append(painter)
      #painting_info.append(pigment_cn)
      #painting_info.append(pigment_en)
      painting_info.append(pigment)
      painting_info.append(tag)
      tag=''
      painting_info.append(attr)
      painting_info.append(img)
      painting_info_list.append(painting_info)
      painting_info=[]
```
